```python
### removing null values via mean , median
```

```python
num_col_null = ['LB',
 'AC',
 'FM',
 'UC',
 'DL',
 'DS',
 'DP',
 'ASTV',
 'MSTV',
 'ALTV',
 'MLTV',
 'Width',
 'Tendency',
 'NSP']
for column in num_col_null:
    mean=df[column].mean()
    df[column].fillna(mean,inplace=True)
```

```python
num_col_null = ['LB',
 'AC',
 'FM',
 'UC',
 'DL',
 'DS',
 'DP',
 'ASTV',
 'MSTV',
 'ALTV',
 'MLTV',
 'Width',
 'Tendency',
 'NSP']
for column in num_col_null:
    median=df[column].median()
    df[column].fillna(median,inplace=True)
```

### Remove outliers

```python
import scipy.stats as stats
from scipy.stats import zscore
```

```python
z_scores = zscore(df.select_dtypes(include=['float64', 'int64']))
threshold = 3
outlier_mask = (abs(z_scores) > threshold).any(axis=1)
df_cleaned = df[~outlier_mask]
```

## 4. Pattern Recognition and Insights

### Plot the Heat Map

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df[num_col].corr(),annot=True,cmap=plt.cm.Blues,fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

### Plot the Pair Plot

```python
plt.figure(figsize=(10,8))
sns.pairplot(df[num_col])
```

```python
num_col = ['LB',
 'AC',
 'FM',
 'UC',
 'DL',
 'DS',
 'DP',
 'ASTV',
 'MSTV',
 'ALTV',
 'MLTV',
 'Width',
 'Tendency',
 'NSP']

correlation_matrix = df[num_col].corr()

# Get strong correlations (between 0.5 and 1.0)
strong_correlations = correlation_matrix[(correlation_matrix > 0.5) & (correlation_matrix < 1.0)]

print("Strong Correlations:")
print(strong_correlations)
```
