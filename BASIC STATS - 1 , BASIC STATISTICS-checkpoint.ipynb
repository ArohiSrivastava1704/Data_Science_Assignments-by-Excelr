### IQR

```python
q1 = df[num_cols].quantile(0.25)
q3 = df[num_cols].quantile(0.75)
iqr = q3-q1
lower_extreme = q1-1.5*iqr
upper_extreme = q3+1.5*iqr
```

### TO CHECK WHICH DATA HAS THE OUTLIERS

```python
import pandas as pd
lower_extreme = pd.Series({
    'Volume': -1.5,
    'Avg Price': -13987.5,
    'Total Sales Value': -73050.0,
    'Discount Rate (%)': 7.740579,
    'Discount Amount': -6823.594880,
    'Net Sales Value': -66266.347664
})

upper_extreme = pd.Series({
    'Volume': 10.5,
    'Avg Price': 24552.5,
    'Total Sales Value': 128950.0,
    'Discount Rate (%)': 24.339202,
    'Discount Amount': 12600.549611,
    'Net Sales Value': 116316.469161
})
```

```python
detect_outliers = ((df[num_cols] < lower_extreme.loc[num_cols]) | (df[num_cols] > upper_extreme.loc[num_cols]))
df_outliers = df[detect_outliers.any(axis=1)][num_cols]
print(df_outliers)
```

```python
pd.DataFrame(df_outliers)
```

## to get the data without outliers

```python
detect_outliers = ((df[num_cols] >= lower_extreme.loc[num_cols]) & (df[num_cols] <= upper_extreme.loc[num_cols]))
df_outliers = df[detect_outliers.any(axis=1)][num_cols]
print(df_outliers)
```

```python
 pd.DataFrame(df_outliers)
```

```python
# Detect non-outliers (values within the lower and upper bounds)
detect_non_outliers = (df[num_cols] >= lower_extreme.loc[num_cols]) & (df[num_cols] <= upper_extreme.loc[num_cols])

# Keep only rows where all columns have non-outlier values
df_non_outliers = df[detect_non_outliers.all(axis=1)][num_cols]

print(df_non_outliers)
```

```python
print(df_non_outliers.describe())
```

## Standardization of Numerical Variables

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.preprocessing import StandardScaler
```

### Standardization, also known as z-score normalization, is a method used to scale numerical variables to have a mean of 0 and a standard deviation of 1. It transforms the data distribution in such a way that the mean of the standardized data is 0 and the standard deviation is 1. This process is particularly useful when the features have different scales or units, making it difficult for certain algorithms to interpret them correctly. The formula for standardization (z-score normalization) is: z= (x-mu)/sigma)

```python
## histogram plot before standardization
```

```python
for columns in num_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[columns], color='Blue')
    plt.xlabel(columns , color='red' , fontsize=14) 
    plt.ylabel('Frequency' , color='green' , fontsize=14) 
    plt.title((f'Distribution of {column} - Before Standardization'),color='brown',fontsize=18)
    plt.show()
```

## Conversion of Categorical Data into Dummy Variables

### Converting categorical data into dummy variables, also known as one-hot encoding, is necessary because many machine learning algorithms cannot directly handle categorical data. One-hot encoding transforms categorical variables into a binary format where each category becomes a binary column (0 or 1)

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
```

```python
df
```

```python
cat_col
```

```python
one_hot=make_pipeline(OneHotEncoder())  # pipeline means taking all the categories together
ord_enc=make_pipeline(OrdinalEncoder())
```

```python
transformers=ColumnTransformer([('one_hot',one_hot,['Date','City']) , ('ord_enc',ord_enc,['Day','SKU','BU','Brand','Model'])]) #1st one_hot is the sting , same in the 2nd case ord_enc is string
transformers
```

```python
transformers.get_feature_names_out
```

```python
df1 = pd.DataFrame(transformers.fit_transform(df[cat_col]), columns=transformers.get_feature_names_out())
df1
```
