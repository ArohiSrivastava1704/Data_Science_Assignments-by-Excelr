```python
from scipy.stats import stats
```

```python
def remove_outlier_z_score(df,columns):
    thershold=2
    z_scores=stats.zscore(df[columns])
    return df[(abs(z_scores)<thershold).all(axis=1)]
```

```python
df_new=remove_outlier_z_score(df,columns)
df_new
```

```python
corr=df_new.corr()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(corr,annot=True,cmap=plt.cm.Blues)
plt.show()
```

# Task 2: Dimensionality Reduction with PCA:

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
std_sca=StandardScaler()
```

```python
scaled=std_sca.fit_transform(df_new)
scaled
```

```python
model_pca=PCA(n_components=10)
model_pca.fit_transform(scaled)
```

```python
pca_transformed_data = model_pca.transform(scaled)  
```

```python
variance = model_pca.explained_variance_ratio_
variance
```

```python
variance.cumsum()
```

# Task 3: Clustering with Original Data:

```python
## k means 
```

```python
from sklearn.cluster import KMeans
```

```python
model_k=KMeans(n_clusters=10)
```

```python
model_k.fit_transform(scaled)
```

```python
model_k.inertia_
```

```python
model_k.labels_
```

```python
df_new['labels']=model_k.labels_
df_new
```

```python
sns.scatterplot(x=pca_transformed_data[:, 0], y=pca_transformed_data[:, 1], hue=model_k.labels_, palette='viridis')
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(scaled,model_k.labels_)
```

# Task 4: Clustering with PCA Data:

```python
pca_transformed_data
```

```python
kmeans_pca = KMeans(n_clusters=3, random_state=42)
pca_clusters = kmeans_pca.fit_predict(pca_transformed_data)
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x=pca_transformed_data[:, 0], y=pca_transformed_data[:, 1], hue=pca_clusters, palette='viridis')
plt.title('Clustering with PCA Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
labels=kmeans_pca.labels_
labels
```

```python
silhouette_score(pca_transformed_data,labels)
```
