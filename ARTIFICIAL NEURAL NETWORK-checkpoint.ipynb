# 2. Model Implementation

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(x, y, train_size=0.2, random_state=42)
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Build the ANN model
model = Sequential()
model.add(Dense(64, input_dim=16, activation='relu'))  # Set input_dim to 16
model.add(Dense(32, activation='relu'))  # Hidden layer
model.add(Dense(len(lab_enc.classes_), activation='softmax'))  # Output layer for multi-class classification
```

```python
# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
model.fit(xtrain, ytrain, epochs=100, batch_size=10, validation_split=0.2)
```

```python
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier
```

```python
def create_model(neurons=32 , optimizer='adam'):
    model = Sequential()
    model.add(Dense(neurons, input_dim=16, activation='relu'))
    model.add(Dense(len(y.unique()), activation='softmax'))
    model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
```

```python
model = KerasClassifier(model=create_model, verbose=0)
```

```python
### Define the grid search parameters
```

```python
param_grid = {
    'model__neurons': [32, 64],  # Use model__ to specify parameters for the model
    'batch_size': [10, 20],
    'epochs': [50, 100],
    'optimizer': ['adam', 'rmsprop']
}
```

```python
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3)
grid_result = grid.fit(xtrain, ytrain)
```

```python
# Print the best parameters
print(f"Best: {grid_result.best_score_} using {grid_result.best_params_}")
```
