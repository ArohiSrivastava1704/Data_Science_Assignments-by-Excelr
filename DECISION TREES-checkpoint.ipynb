# Interview Questions

## 1. What are some common hyperparameters of decision tree models, and how do they affect the model's performance?

1. Max_depth: Controls the maximum depth of the tree , Shallow trees may underfit; deep trees may overfit.

2. Min_samples_split: Minimum samples required to split an internal node , Higher values reduce overfitting; lower values can lead to overfitting.

3. Min_samples_leaf: Minimum samples required to be at a leaf node , Higher values reduce overfitting; lower values may overfit.

4. Max_features: Number of features to consider for splitting , Smaller values prevent overfitting but may reduce accuracy.

5. Max_leaf_nodes: Limits the number of leaf nodes , Fewer nodes reduce overfitting, but too few may decrease accuracy.

6. Criterion: Measures the quality of a split ('gini' or 'entropy') , 'gini' is faster, 'entropy' is slightly more accurate but computationally expensive.

## 2. What is the difference between the Label encoding and One-hot encoding?

1. Label Encoding: Each category is assigned a unique integer (e.g., "red" = 0, "blue" = 1, "green" = 2). 

   -> It works well for ordinal data (like 'low', 'medium', 'high'), where the order matters. However, it may create problems for models that treat the encoded numbers as having an inherent relationship or ranking, which doesn't exist in nominal data (e.g., "red" is not less than "blue").

2. One-Hot Encoding: Each category is represented as a binary vector where only one element is 1, and the rest are 0 (e.g., "red" = [1, 0, 0], "blue" = [0, 1, 0], "green" = [0, 0, 1]). 

   -> This avoids any ordinal assumptions and is ideal for nominal data, where no natural ordering exists. However, it increases the feature space, which can lead to higher memory usage and computational complexity.
